```python
import pandas as pd
```

```python
data_path = 'yellow_tripdata_2024-01.parquet'
data = pd.read_parquet(data_path)
data.head()
```

```python
data.tail()
```

```python
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.boxplot(data["trip_distance"])
plt.title("trip_distance")

plt.subplot(1, 2, 2)
plt.boxplot(data["total_amount"])
plt.title("total_amount")

plt.tight_layout()
plt.show()
```

```python
# passenger_count : The number of passengers in the vehicle (This is a driver-entered value)

# RatecodeID : The final rate code in effect at the end of the trip (1= Standard rate, 2=JFK, 3=Newark, 4=Nassau or Westchester,
# 5=Negotiated fare, 6=Group ride)

# store_and_fwd_flag : This flag indicates whether the trip record was held in vehicle memory before sending to the vendor, aka store and forward,
# because the vehicle did not have a connection to the server.(Y= store and forward trip, N= not a store and forward trip)

# Congestion_Surcharge : Total amount collected in trip for NYS congestion surcharge.

# Airport_fee : $1.25 for pick up only at LaGuardia and John F. Kennedy Airports
```

```python
data.shape
```

```python
# Filtrage et affichage de 2-3 lignes avec des valeurs manquantes
print("\nExemples de lignes avec des valeurs manquantes:")
print(data[data.isnull().any(axis=1)].head(3))
```

```python
# Conversion des colonnes datetime
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])

# Calcul du nombre total d'enregistrements par jour
total_pickups_per_day = data['tpep_pickup_datetime'].dt.date.value_counts().sort_index()

# Calcul des valeurs manquantes par jour pour les colonnes spcifiques
missing_data_pickup = data[data[['passenger_count', 'RatecodeID', 'store_and_fwd_flag', 'congestion_surcharge', 'Airport_fee']].isnull().any(axis=1)]['tpep_pickup_datetime'].dt.date.value_counts().sort_index()
```

```python
from tabulate import tabulate
results = pd.DataFrame({
    "Total Pickups": total_pickups_per_day,
    "Missing Pickups": missing_data_pickup,
    "Proportion Missing Pickups (%)": (missing_data_pickup / total_pickups_per_day * 100).round(2),
})

results.fillna(0, inplace=True)

print("Rsum des valeurs manquantes par jour pour les pickups :")
print(tabulate(results, headers='keys', tablefmt='psql', showindex=True))
```
